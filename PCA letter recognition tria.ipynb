```python
def img_to_vector(filename):
    img  = Image.open(filename).resize((112, 92)) 
    img_vector = np.array(img).flatten()
    sz = img_vector.shape[0]
    #img_vector = img_vector.reshape(sz,1)
    return img_vector
```

```python
def preprocess():
    train_set = []
    train_set_number = []
    test_set = []
    test_set_number = []
    counter = 0
    
    for i in range(840):
        person_id = (i%28)+1
        for j in range(8):
                counter = counter + 1
                print(person_id,counter)
                path =r'D:\Collage\TERM 4\Arabic handwritten characters\Train Images 13440x32x32\train\id_'+str(counter)+'_label_'+str(person_id)+'.png'
                img = img_to_vector(path).astype(np.int64) 
                train_set.append(img)
                train_set_number.append(person_id)
    counter = 0
    for i in range(140):
        person_id = (i%28)+1
        for j in range(2):
                counter = counter + 1
                print(person_id,counter)
                path =r'D:\Collage\TERM 4\Arabic handwritten characters\Test Images 3360x32x32\test\id_'+str(counter)+'_label_'+str(person_id)+'.png'
                img = img_to_vector(path).astype(np.int64)
                test_set.append(img)
                test_set_number.append(person_id)
        
        
    train_set = np.array(train_set)
    train_set_number = np.array(train_set_number)
    test_set = np.array(test_set)
    test_set_number = np.array(test_set_number)
    return train_set.T, train_set_number, test_set.T, test_set_number
```

```python
def find_mean(arr):
    M = np.mean(arr, axis = 1)
    return M
```

```python
M = find_mean(train_set)
print(M)
```

```python
def mean_normalization(arr):
    M = find_mean(arr)
    x_range = arr.shape[0]
    y_range = arr.shape[1]
    for i in range(x_range):
        for j in range(y_range):
            arr[i][j] -= M[i]
    return arr 
```

```python
A = mean_normalization(train_set)
A = A
print(A)
```

```python
print(A.shape)
```

```python
L = np.dot(A.T, A)
print(L.shape)
print(L)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(L)
# sort eigenvectors by the value of eigenvalues
print("eigenvalues shape : ", eigenvalues.shape)
print("eigenvalues : ", eigenvalues)
print("eigenvectors shape : ", eigenvectors.shape)
print("eigenvectors : ", eigenvectors)
```

```python
U = []
sz = L.shape[0]
for i in range(sz):
    U.append(np.dot(A,eigenvectors[:,i]))

U = np.array(U)
U = U.T
print(U.shape)
print(U)
```

```python
weight_vector = np.dot(U.T, A)
print(weight_vector.shape)
print(weight_vector)
```
